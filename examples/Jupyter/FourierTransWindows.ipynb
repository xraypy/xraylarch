A demonstration of XAFS Fourier transform windows and settings.

Feel free to explore and play with options 


```python
import numpy as np
from larch.xafs import ftwindow
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 401)
```

```python
# show Hanning window and "sills"

hann_win1 =  ftwindow(x, xmin=5, xmax=15, dx=3, window='hanning')

plt.plot(x, hann_win1, label='Hanning(5, 15, dx=3)')

plt.plot([3.5, 3.5], [0, 1],   'k--')
plt.plot([6.5, 6.5], [0, 1],   'k--')
plt.plot([5.0, 5.0], [0, 0.5], 'k--') 
plt.text(3.75, 1.05, 'dx=3')
plt.arrow(5.0, 1, -1.3, 0, width=0.01, head_length=0.2)
plt.arrow(5.0, 1,  1.3, 0, width=0.01, head_length=0.2)
plt.xlabel('x')
plt.ylim(-0.2, 1.3)
plt.legend()
```

```python
# now compare some different window functions

hann_win2 =  ftwindow(x, xmin=5, xmax=15, dx=5, window='hanning')
hann_win3 =  ftwindow(x, xmin=5, xmax=15, dx=3, dx2=7, window='hanning')
parzen_win1 =  ftwindow(x, xmin=5, xmax=15, dx=3, window='parzen')
welch_win1 =  ftwindow(x, xmin=5, xmax=15, dx=3, window='welch')

plt.plot(x, hann_win1, label='Hanning(5, 15, dx=3)')
plt.plot(x, hann_win2, label='Hanning(5, 15, dx=5)')
plt.plot(x, parzen_win1, label='Parzen(5, 15, dx=3)')
plt.plot(x, welch_win1, label='Welch(5, 15, dx=3)')

plt.xlabel('x')
plt.ylim(-0.1, 1.05)
plt.legend()
```

```python
# compare setting different "dx" values for one side 

plt.plot(x, hann_win1, label='Hanning(5, 15, dx=3)')
plt.plot(x, hann_win3, label='Hanning(5, 15, dx=3, dx2=7)')

plt.xlabel('x')
plt.ylim(-0.1, 1.05)
plt.legend()
```

```python
# compare different functional forms

kai_win1 =  ftwindow(x, xmin=5, xmax=15, dx=4, window='kaiser')
sin_win1 =  ftwindow(x, xmin=5, xmax=15, dx=4, window='sine')
gau_win1 =  ftwindow(x, xmin=5, xmax=15, dx=4, window='gaussian')

plt.plot(x, kai_win1, label='Kaiser(5, 15, dx=4)')
plt.plot(x, sin_win1, label='Sine(5, 15, dx=4)')
plt.plot(x, gau_win1, label='Gassian(5, 15, dx=4)')

plt.plot(x, kai_win1, label='Kaiser(5, 15, dx=4)')
plt.plot(x, ftwindow(x, xmin=5, xmax=15, dx=10, window='kaiser'),   label='Kaiser(5, 15, dx=10)')
plt.plot(x, ftwindow(x, xmin=5, xmax=15, dx=3, window='gaussian'),  label='Gaussian(5, 15, dx=3)')

plt.xlabel('x')
plt.ylim(-0.1, 1.05)
plt.legend()
```
