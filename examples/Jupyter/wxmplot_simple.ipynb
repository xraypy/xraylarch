`````python
# this example demonstrates using wxmpot from a Jupyter notebook

import numpy as np
x = np.linspace(-10, 10, 251)
y = np.sin(x) + x/100

import wxmplot.interactive as wi

wi.plot(x, y, label='sin+x', theme='dark', xlabel=r'$x\, (\mathrm{{\mu}m})$', new=True)
`````

`````python
wi.plot(x, np.cos(x*1.1), label='cos')
`````


The plot will be rendered in an external wx panel, outside of the Jupyter notebook, and will be fully interactive: zoom, pan, adjust the colors, text, and so forth.  But it will not be shown within the Jupyter notebook itself.
